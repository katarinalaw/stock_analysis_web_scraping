```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://stockanalysis.com/list/biggest-companies/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup)
```

```python
soup.find('table', class_ ='svelte-1tv1ofl')
```

```python
stock_titles = soup.find_all('th')

stock_titles
```

```python
#getting the title and removing unneccessary empty space for each title
table_titles = [title.text.strip() for title in stock_titles]

print(table_titles)
```

```python
import pandas as pd
```

```python
#creating dataframe
df = pd.DataFrame(columns = table_titles)

df
```

```python
#tr is row data
column_data = soup.find_all('tr')
```

```python
#td is individual data
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_data = [data.text.strip() for data in row_data]
    print(individual_data)
    
    #checking the length of the dataframe every time it's looping through
    df_length = len(df)
    #putting the info into the next position
    df.loc[df_length] = individual_data
```

```python
#this shows the complete dataframe including all rows info
df
```
